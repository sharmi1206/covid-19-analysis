```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_zones['Zone'] = le.fit_transform(df_zones['Zone'])
df_zones.head()  #Red -> 3, Green -> 1, Orange -> 2
```

```python
state_zones = df_zones.groupby(['State', 'Zone']).size().reset_index(name='counts')
state_zones
```

# Label Encoding of Zones

```python
#Red -> 3, Green -> 1, Orange -> 2
stateRedZoneMap ={}
stateGreenZoneMap ={}
stateOrangeZoneMap ={}
for j in range(0, len(state_zones)):
    stateName = state_zones.iloc[j]['State']
    stateRedZoneMap[stateName] = 0
    stateGreenZoneMap[stateName] = 0
    stateOrangeZoneMap[stateName] = 0


for j in range(0, len(state_zones)):
    stateName = state_zones.iloc[j]['State']
    zoneCount = state_zones.iloc[j]['counts']
    if(state_zones.iloc[j]['Zone'] == 1):
        stateGreenZoneMap[stateName] = zoneCount
    elif(state_zones.iloc[j]['Zone'] == 2):
        stateOrangeZoneMap[stateName] = zoneCount
    elif(state_zones.iloc[j]['Zone'] == 3):
        stateRedZoneMap[stateName] = zoneCount
```
